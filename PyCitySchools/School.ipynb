# PyCitySchool

```python
import pandas as pd
import numpy as np
import os

#loading files
school_datafile = os.path.join('Resources', 'schools_complete.csv')
student_datafile = os.path.join('Resources', 'students_complete.csv')

school_data = pd.read_csv(school_datafile)
student_data = pd.read_csv(student_datafile)

  
school_data_merged = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_merged
```

## DISTRICT SUMMARY

```python
total_schools = len(school_data_merged["school_name"].unique())
#total_schools
total_students = (school_data_merged["student_name"].count())
#total_students
Total_budget = sum(school_data_merged["budget"].unique())
#Total_budget
average_math_score = school_data_merged["math_score"].mean()
#average_math_score
average_reading_score = school_data_merged["reading_score"].mean()
#average_reading_score
table_of_kiddos_who_passed_maths = school_data_merged.loc[school_data_merged["math_score"]>=70] #table_of_kiddos_who_passed
number_of_maths_passers = table_of_kiddos_who_passed_maths["math_score"].count()
avergae_score_of_maths_passers = number_of_maths_passers/total_students*100
#avergae_score_of_maths_passers
table_of_kiddos_who_passed_reading = school_data_merged.loc[school_data_merged["reading_score"]>=70] #table_of_kiddos_who_passed
number_of_reading_passers = table_of_kiddos_who_passed_reading["reading_score"].count()
avergae_score_of_reading_passers = number_of_reading_passers/total_students*100
#avergae_score_of_reading_passers
overall_passed = school_data_merged.loc[(school_data_merged['math_score']>=70) 
                                          & (school_data_merged['reading_score']>=70)]
stds_overall_passed = overall_passed['reading_score'].count()
percent_overall_passed = stds_overall_passed / total_students * 100
percent_overall_passed
# Create a dictionary to hold the above results
district_summary = {
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [avg_math_score],
    'Average Reading Score' : [avg_reading_score],
    '% Passing Math': [percent_math_passed],
    '% Passing Reading': [percent_reading_passed],
    '% Overall Passing': [percent_overall_passed]
}

#dataframe
district_summary_df = pd.DataFrame(district_summary)
district_summary_df['Total Students'] = district_summary_df['Total Students'].map('{:,}'.format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,.2f}'.format)
district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map('{:.4f}'.format)   
district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map('{:.4f}'.format)   
district_summary_df['% Passing Math'] = district_summary_df['% Passing Math'].map('{:.4f}'.format)
district_summary_df['% Passing Reading'] = district_summary_df['% Passing Reading'].map('{:.4f}'.format)
district_summary_df['% Overall Passing'] = district_summary_df['% Overall Passing'].map('{:.4f}'.format)

district_summary_df
```

## School Summary

```python
# School Names (Grouping by School Names, School names become index of the dataframe)
schools_grouped = school_data_merged.groupby(['school_name'])

# School Types (Use df.first() to pick up data that is found first)
school_types = schools_grouped['type'].first()

# Total Students (Use df.count() to count # of students per school in grouped dataframe)
total_std = schools_grouped['student_name'].count()

# Total School Budget
school_budget = schools_grouped['budget'].first()

# Per Student Budget
per_std_bg = school_budget / total_std

# Average Math Score (Use df.mean() to find an average of numeric data)
avg_math_school = schools_grouped['math_score'].mean()

# Average Reading Score
avg_reading_school = schools_grouped['reading_score'].mean()

# % Passing Math (Use condition to filter only needed data rows)
math_pass_df = school_data_merged[school_data_merged['math_score'] >= 70]
math_pass_gp = math_pass_df.groupby(['school_name'])
percent_pass_math = math_pass_gp['math_score'].count() / total_std *100

# % Passing Reading
reading_pass_df = school_data_merged[school_data_merged['reading_score'] >= 70]
reading_pass_gp = reading_pass_df.groupby(['school_name'])
percent_pass_reading = reading_pass_gp['reading_score'].count() / total_std *100

# % Overall Passing (The percentage of students that passed math and reading.)
overall_pass_df = school_data_merged[(school_data_merged['math_score'] >= 70) & (school_data_merged['reading_score'] >= 70)]
overall_pass_gp = overall_pass_df.groupby(['school_name'])
percent_pass_overall = overall_pass_gp['student_name'].count() / total_std *100

# Create a dictionary to hold the above results
schools_summary_dict = {
    'School Type': school_types,
    'Total Students' : total_std,
    'Total School Budget': school_budget,
    'Per Student Budget': per_std_bg,
    'Average Math Score': avg_math_school,
    'Average Reading Score': avg_reading_school,
    '% Passing Math': percent_pass_math,
    '% Passing Reading': percent_pass_reading,
    '% Overall Passing': percent_pass_overall 
}

# Create a dataframe to hold a dictionary
city_schools_summary_df = pd.DataFrame(schools_summary_dict)

# Give the displayed data cleaner formatting
schools_summary_df = city_schools_summary_df.copy()
schools_summary_df.index.name = 'School Name'
schools_summary_df['Total School Budget'] = schools_summary_df['Total School Budget'].map('${:,}'.format)
schools_summary_df['Per Student Budget'] = schools_summary_df['Per Student Budget'].map('${:,}'.format)

# Present the data
schools_summary_df
```

## Scores by Different Categories
Scores by School Spending

```python
# Create bins for segmenting and sorting data
spending_bins = [0, 584, 629, 644, 675]

# Define name for each data that falls in each interval index(bin) 
spending_group_name = ['<$584', '$585-629', '$630-644', '$645-675']

# Categorize data
# * Adding another column to identify data by bins
# * Apply pd.cut() to segment and sort data values into bins 
#   (include_lowest=True > The first interval will be inclusive) 
city_schools_summary_df['Spending Ranges (Per Student)'] = pd.cut(city_schools_summary_df['Per Student Budget'], 
                                                                  spending_bins, 
                                                                  labels=spending_group_name, 
                                                                  include_lowest=True)

# Grouping dataframe based on Spending Ranges (Per Student)
school_spending_grouped = city_schools_summary_df.groupby(['Spending Ranges (Per Student)'])

# Calculate the average scores by school spending per student
math_by_spending = school_spending_grouped['Average Math Score'].mean()
reading_by_spending = school_spending_grouped['Average Reading Score'].mean()
percent_math_spending = school_spending_grouped['% Passing Math'].mean()
percent_reading_spending = school_spending_grouped['% Passing Reading'].mean()
percent_overall_spending = school_spending_grouped['% Overall Passing'].mean()

# Create the dictionary to hold the results
scores_by_school_spending_dict = {
    'Average Math Score': math_by_spending,
    'Average Reading Score': reading_by_spending,
    '% Passing Math': percent_math_spending,
    '% Passing Reading': percent_reading_spending,
    '% Overall Passing': percent_overall_spending
}

# Create the dataframe to hold the result dictionary
scores_by_school_spending_df = pd.DataFrame(scores_by_school_spending_dict)

# Present dataframe
scores_by_school_spending_df
```

Scores by School Type

```python
# Create bins for segmenting and sorting data
size_bins = [0, 999, 1999, 5000]

# Define name for each data that falls in each interval index(bin) 
school_size_name = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']

# Categorize data
# * Adding another column to identify data by bins
# * Apply pd.cut() to segment and sort data values into bins 
#   (include_lowest=True > The first interval will be inclusive)
city_schools_summary_df['School Size'] = pd.cut(city_schools_summary_df['Total Students'], 
                                                size_bins, 
                                                labels=school_size_name, 
                                                include_lowest=True)

# Grouping dataframe based on School Size
school_size_grouped = city_schools_summary_df.groupby(['School Size'])

# Calculate the average scores by school size
math_by_size = school_size_grouped['Average Math Score'].mean()
reading_by_size = school_size_grouped['Average Reading Score'].mean()
percent_math_size = school_size_grouped['% Passing Math'].mean()
percent_reading_size = school_size_grouped['% Passing Reading'].mean()
percent_overall_size = school_size_grouped['% Overall Passing'].mean()

# Create the dictionary to hold the results
scores_by_school_size_dict = {
    'Average Math Score': math_by_size,
    'Average Reading Score': reading_by_size,
    '% Passing Math': percent_math_size,
    '% Passing Reading': percent_reading_size,
    '% Overall Passing': percent_overall_size
}

# Create the dataframe to hold the result dictionary
scores_by_school_size_df = pd.DataFrame(scores_by_school_size_dict)

# Present dataframe
scores_by_school_size_df
```
